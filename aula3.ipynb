```python
import numpy as np
import matplotlib.pyplot as plt
# Função para calcular a translação dos pontos
def translacao(pontos, Tx, Ty):
  pontos_transladados = []
  for ponto in pontos:
    x_u = ponto[0] + Tx
    y_u = ponto[1] + Ty
    pontos_transladados.append((x_u, y_u))
  return pontos_transladados
# Pontos originais
p1 = (0, 0)
p2 = (2, 2)
# Vetor de translação
Tx = 3
Ty = 2
# Calcular a translação dos pontos
pontos_transladados = translacao([p1, p2], Tx, Ty)
# Plotar os pontos originais e os pontos transladados
plt.xlim(-1, 5.5)
plt.ylim(-1, 5.5)
plt.plot([p1[0], p2[0]], [p1[1], p2[1]], 'bo-', label='Pontos originais')
plt.plot([ponto[0] for ponto in pontos_transladados], [ponto[1] for ponto in pontos_transladados],'ro-', label='Pontos transladados')
#[ponto[0] for ponto in pontos_transladados]:
#Isso é uma compreensão de lista, uma construção Python que permite criar listas de maneira mais
# compacta.
#O que está entre os colchetes é uma expressão que define como cada elemento da lista será
# construído.
#ponto[0] é a expressão usada para cada elemento da lista. Aqui, 'ponto' representa cada elemento
# da lista 'pontos_transladados', que são as coordenadas dos pontos transladados.
#ponto[0] extrai o primeiro elemento de cada tupla dentro da lista pontos_transladados, que
# corresponde à coordenada x do ponto.
#Então, [ponto[0] for ponto in pontos_transladados] cria uma lista contendo apenas as coordenadas
# x dos pontos transladados. Essa lista é usada para definir os valores do eixo x no gráfico.
#'bo-' especifica o estilo do gráfico: b para azul (blue), 'o' para marcador de círculo e '-' para linha
# contínua.
#'ro-':#Isso define o estilo do gráfico para os pontos transladados. Aqui, 'r' indica que os pontos
# serão vermelhos,
# 'o' especifica que os marcadores serão círculos e '-' indica que uma linha contínua conectará os
# pontos.
# Configurações do gráfico
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Translação de pontos no plano cartesiano')
plt.grid(True)
plt.legend()
# Mostrar o gráfico
plt.show()
```

```python
p1 = (6,8)
p2 = (4,5)
p3 = (8,5)
# Vetor de translação
Tx = 3
Ty = -4
# Calcular a translação dos pontos
pontos_transladados = translacao([p1, p2, p3], Tx, Ty)
pontos_transladados.append(pontos_transladados[0])
# Plotar os pontos originais e os pontos transladados
plt.xlim(2, 12)
plt.ylim(0, 12)
plt.plot([p1[0], p2[0], p3[0], p1[0]], [p1[1], p2[1], p3[1], p1[1]], 'bo-', label='Pontos originais')
plt.plot([ponto[0] for ponto in pontos_transladados], [ponto[1] for ponto in pontos_transladados],'ro-', label='Pontos transladados')

for i in range(len(pontos_transladados)):
    print(f"Ponto {i+1}: ({pontos_transladados[i][0]}, {pontos_transladados[i][1]})")

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Translação de pontos no plano cartesiano')
plt.grid(True)
plt.legend()

plt.show()
```

```python
def escala(pontos, Sx, Sy):
  pontos_transladados = []
  for ponto in pontos:
    x_u = ponto[0] * Sx
    y_u = ponto[1] * Sy
    pontos_transladados.append((x_u, y_u))
  return pontos_transladados
p1 = (-1,-1)
p2 = (1, -1)
p3 = (1, 1)
p4 = (-1, 1)
Sx = 2
Sy = 2
pontos_escalados = escala([p1, p2, p3, p4], Sx, Sy)
pontos_escalados.append(pontos_escalados[0])
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.plot([p1[0], p2[0], p3[0], p4[0], p1[0]], [p1[1], p2[1], p3[1], p4[1], p1[1]], 'bo-', label='Pontos originais')
plt.plot([ponto[0] for ponto in pontos_escalados], [ponto[1] for ponto in pontos_escalados],'ro-', label='Pontos transladados')

for i in range(len(pontos_escalados)):
    print(f"Ponto {i+1}: ({pontos_escalados[i][0]}, {pontos_escalados[i][1]})")

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Translação de pontos no plano cartesiano')
plt.grid(True)
plt.legend()

plt.show()
```

```python
def rotacao(pontos, angulo):
  angulo = np.radians(angulo)
  pontos_rotacao = []
  for ponto in pontos:
    xu = ponto[0] * np.cos(angulo) - ponto[1] * np.sin(angulo)
    yu = ponto[1] * np.cos(angulo) + ponto[0] * np.sin(angulo)
    pontos_rotacao.append((xu, yu))
  return pontos_rotacao

grau = int(input('Digite o ângulo de rotação: '))
pontos_rotacao = rotacao([p1, p2, p3, p4], grau)
pontos_rotacao.append(pontos_rotacao[0])
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.plot([p1[0], p2[0], p3[0], p4[0], p1[0]], [p1[1], p2[1], p3[1], p4[1], p1[1]], 'bo-', label='Pontos originais')
plt.plot([ponto[0] for ponto in pontos_rotacao], [ponto[1] for ponto in pontos_rotacao],'ro-', label='Pontos rotacionado')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def hermite_curve(P1, P2, M1, M2, t):
    h1 = 2*t**3 - 3*t**2 + 1
    h2 = t**3 - 2*t**2 + t
    h3 = -2*t**3 + 3*t**2
    h4 = t**3 - t**2

    x = h1*P1[0] + h2*M1[0] + h3*P2[0] + h4*M2[0]
    y = h1*P1[1] + h2*M1[1] + h3*P2[1] + h4*M2[1]

    return x, y


P1 = np.array([0, 0])
P2 = np.array([3, 3])
M1 = np.array([6, -6])
M2 = np.array([6, 4])


t_values = np.linspace(0, 1, 50)

x_values, y_values = [], []
for t in t_values:
    x, y = hermite_curve(P1, P2, M1, M2, t)
    x_values.append(x)
    y_values.append(y)

plt.plot(x_values, y_values, label="Curva de Hermite", color="blue")
plt.scatter([P1[0], P2[0]], [P1[1], P2[1]], color="red", label="Pontos de Controle")
plt.quiver(P1[0], P1[1], M1[0], M1[1], angles='xy', scale_units='xy', scale=1, color="green", label="Derivada M1")
plt.quiver(P2[0], P2[1], M2[0], M2[1], angles='xy', scale_units='xy', scale=1, color="orange", label="Derivada M2")
plt.legend()
plt.title("Curva de Hermite")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```
